## Accessing the data

```python
f=open('groceries.csv','r')
f
```

```python
content=f.readlines()
content
```

```python
type(content)
```

```python
len(content)
```

```python
items=[]
for x in content:
    item=x.strip()
    items.append(item.split(','))
items
```

```python
from mlxtend.preprocessing import TransactionEncoder

te=TransactionEncoder()
items_encoded=te.fit_transform(items)
items_encoded
```

len(items_encoded)

```python
import pandas as pd
import numpy as np
```

```python
# COnverting to Pandas DF

items_df=pd.DataFrame(items_encoded, columns=te.columns_)
items_df
```

## Building the model

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_set=apriori(items_df,min_support=0.01,use_colnames=True)
frequent_set
```

```python
rules=association_rules(frequent_set,metric='lift')
rules
```

```python
rules.sort_values('confidence',ascending= False)
```
